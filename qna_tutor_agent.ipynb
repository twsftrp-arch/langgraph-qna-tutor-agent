# Trinity RAG Assistant

Education Agent Advanced Features + Streamlit 미션 제출용 LangGraph 노트북입니다.

## 핵심 구조
- Supervisor가 질문 의도를 분류해 Tutor, Quiz, Researcher 전문 에이전트로 라우팅합니다.
- Supervisor는 질문과 학생 권한에 맞춰 검색 worker를 동적으로 만들고, LangGraph `Send` API로 병렬 실행합니다.
- `data/` 폴더의 실제 Markdown/JSON 자료를 Tool로 검색해 RAG 컨텍스트를 만듭니다.
- Judge Agent는 외부 AI 평가 또는 로컬 fallback rubric으로 초안을 평가합니다.
- `MemorySaver` 체크포인터로 같은 `thread_id`의 질문과 진행 중인 퀴즈를 유지합니다.
- `interrupt()`로 선생님 검토에서 실제로 멈추고, `Command(resume=...)`로 승인 또는 수정한 뒤 재개합니다.
- `streamlit_app.py`는 위 그래프를 기본 채팅 UI로 제공합니다.


## 그래프 구조

```mermaid
flowchart TD
    START --> CheckAuth
    CheckAuth --> LoadMemory
    LoadMemory --> Supervisor
    Supervisor -- Send 1..3 --> ResearchWorkers
    ResearchWorkers --> ResearchSynthesis
    ResearchSynthesis -- 설명 --> Tutor
    ResearchSynthesis -- 문제·채점 --> Quiz
    ResearchSynthesis -- 자료 조사 --> Researcher
    Tutor --> Judge
    Quiz --> Judge
    Researcher --> Judge
    Judge -- 자동 발행 --> Publish
    Judge -- 교사 검토 --> TeacherReview[TeacherReview interrupt]
    TeacherReview -- 승인 --> Publish
    TeacherReview -- 수정 요청 --> ReviseDraft
    ReviseDraft --> TeacherReview
    Publish --> END
```
