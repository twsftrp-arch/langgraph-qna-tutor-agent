# Trinity RAG Assistant

Education Agent 미션 제출용 LangGraph 노트북입니다.

## 핵심 구조
- 학생 권한(`is_enrolled`)에 따라 Premium/Public 경로로 분기합니다.
- 각 경로는 LangGraph `Send` API로 여러 검색 작업을 병렬 실행합니다.
- `data/` 폴더의 실제 Markdown/JSON 자료를 Tool로 검색해 RAG 컨텍스트를 만듭니다.
- `MemorySaver` 체크포인터로 같은 `thread_id`의 질문 기록을 유지합니다.
- 선생님 피드백이 있으면 수정 노드로 돌아간 뒤 다시 리뷰합니다.


## 그래프 구조

```mermaid
graph TD
    START --> CheckAuth
    CheckAuth --> LoadMemory
    LoadMemory -- 수강생 --> PremiumPlan
    LoadMemory -- 비수강생 --> PublicPlan
    PremiumPlan -- Send API --> RetrieveContext
    PublicPlan -- Send API --> RetrieveContext
    RetrieveContext --> DraftAnswer
    DraftAnswer --> TeacherReview
    TeacherReview -- 승인 --> Publish
    TeacherReview -- 수정 요청 --> ReviseDraft
    ReviseDraft --> TeacherReview
    Publish --> END
```
